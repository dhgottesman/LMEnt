```python
import json
import gzip
import pandas as pd

# Read jsonl file (output of wikipedia_parser.py)
file_path = '/home/morg/dataset/parsed_wiki/enwiki-latest-pages-articles17.xml-p20570393p22070392_0.jsonl.gz'
with gzip.open(file_path, 'r') as fin:        
    json_bytes = fin.read()                  

json_str = json_bytes.decode('utf-8')            
data = pd.read_json(path_or_buf=json_str, lines=True)
```

```python
line = data.iloc[0]
```

```python
for entity in line['entities']:
    print(line['text'][entity['entity_start']:entity['entity_end']], ' --- ', entity['entity_name'])
```
