## Example use of tools in the entity linking file:

```python
articles_directory = 'test_dir'
```

```python
if not os.path.isdir(articles_directory):
    print(f"Error: Directory '{articles_directory}' not found.")
else:
    entity_mentions, article_links, stats = process_directory(articles_directory)

    print("Comprehensive Entity Mentions:")
    for article, mentions in entity_mentions.items():
        print(f"  {article}:")
        for mention in mentions:
            print(f"    Span: {mention['span']}, Entity: {mention['entity_title']}")

    print("\nLinks per Article:")
    for article, links in article_links.items():
        print(f"  {article}: {links}")

    print("\nStatistics:")
    print(f"  Total links found by Tool 1: {stats['total_links_tool1']}")
    print(f"  Total links found by Tool 2: {stats['total_links_tool2']}")
    print(f"  Number of links found by both tools: {stats['common_links']}")
    print(f"  Recall of Tool 2 (on Tool 1's links): {stats['recall_tool2']:.4f}")
    print(f"  Precision of Tool 2 (on Tool 1's links): {stats['precision_tool2']:.4f}")
```
